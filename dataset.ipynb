```python
# Import pandas for dataframe handling
import pandas as pd

# Import numpy for random number generation and numeric operations
import numpy as np

# Import os for folder creation and file saving
import os

# Import random for random choices
import random

# Import timedelta and datetime for working with dates
from datetime import datetime, timedelta
```

```python
# Create folder to store uncleaned raw data
os.makedirs("uncleaned_data", exist_ok=True)

# Create folder to store cleaned data later
os.makedirs("data", exist_ok=True)

# Create folder to store trained models later
os.makedirs("models", exist_ok=True)

# Print confirmation
print("Folders created successfully.")
```

```python
# Fix random seed so that output stays reproducible
np.random.seed(42)

# Fix python random seed also
random.seed(42)

# Print confirmation
print("Random seeds set.")
```

```python
# Total number of customer rows
n_customers = 5000

# Create customer ids
customer_ids = [f"CUST_{1000+i}" for i in range(n_customers)]

# Possible gender values
genders = ["Male", "Female", "Other"]

# Possible city values
cities = ["Chennai", "Bangalore", "Hyderabad", "Mumbai", "Delhi", "Pune", "Kolkata"]

# Possible membership values
membership_types = ["Silver", "Gold", "Platinum"]

# Build customer dataframe
customers = pd.DataFrame({
    "customer_id": customer_ids,
    "customer_name": [f"Customer_{i}" for i in range(n_customers)],
    "age": np.random.randint(18, 70, n_customers),
    "gender": np.random.choice(genders, n_customers),
    "city": np.random.choice(cities, n_customers),
    "membership_type": np.random.choice(membership_types, n_customers),
    "join_date": [datetime(2021, 1, 1) + timedelta(days=np.random.randint(0, 1200)) for _ in range(n_customers)]
})

# Show first 5 rows
customers.head()
```

```python
# Add some missing values in age column
customers.loc[np.random.choice(customers.index, 120, replace=False), "age"] = np.nan

# Add some missing values in gender column
customers.loc[np.random.choice(customers.index, 100, replace=False), "gender"] = np.nan

# Add extra spaces and inconsistent casing in city column
customers.loc[np.random.choice(customers.index, 200, replace=False), "city"] = "  chennai  "
customers.loc[np.random.choice(customers.index, 150, replace=False), "city"] = "BANGALORE"

# Duplicate some rows to make data uncleaned
customer_duplicates = customers.sample(40, random_state=42)

# Append duplicates to the original dataframe
customers_uncleaned = pd.concat([customers, customer_duplicates], ignore_index=True)

# Show shape
print("Customers uncleaned shape:", customers_uncleaned.shape)
```

```python
# Total number of product rows
n_products = 5000

# Possible product categories
categories = ["Electronics", "Fashion", "Home", "Beauty", "Groceries"]

# Possible brands
brands = ["BrandA", "BrandB", "BrandC", "BrandD", "BrandE"]

# Build product dataframe
products = pd.DataFrame({
    "product_id": [f"PROD_{2000+i}" for i in range(n_products)],
    "product_name": [f"Product_{i}" for i in range(n_products)],
    "category": np.random.choice(categories, n_products),
    "brand": np.random.choice(brands, n_products),
    "price": np.round(np.random.uniform(100, 5000, n_products), 2),
    "stock_quantity": np.random.randint(10, 500, n_products)
})

# Show first 5 rows
products.head()
```

```python
# Add missing values in brand
products.loc[np.random.choice(products.index, 110, replace=False), "brand"] = np.nan

# Add missing values in price
products.loc[np.random.choice(products.index, 90, replace=False), "price"] = np.nan

# Add inconsistent text
products.loc[np.random.choice(products.index, 160, replace=False), "category"] = " electronics "
products.loc[np.random.choice(products.index, 140, replace=False), "brand"] = "branda"

# Duplicate some rows
product_duplicates = products.sample(45, random_state=42)

# Create uncleaned product dataframe
products_uncleaned = pd.concat([products, product_duplicates], ignore_index=True)

# Show shape
print("Products uncleaned shape:", products_uncleaned.shape)
```

```python
# Total number of order rows
n_orders = 5000

# Possible payment methods
payment_methods = ["UPI", "Card", "Cash on Delivery", "Net Banking"]

# Possible order statuses
order_statuses = ["Delivered", "Cancelled", "Returned", "Shipped"]

# Build orders dataframe
orders = pd.DataFrame({
    "order_id": [f"ORD_{3000+i}" for i in range(n_orders)],
    "customer_id": np.random.choice(customers["customer_id"], n_orders),
    "product_id": np.random.choice(products["product_id"], n_orders),
    "order_date": [datetime(2023, 1, 1) + timedelta(days=np.random.randint(0, 730)) for _ in range(n_orders)],
    "quantity": np.random.randint(1, 6, n_orders),
    "payment_method": np.random.choice(payment_methods, n_orders),
    "order_status": np.random.choice(order_statuses, n_orders, p=[0.70, 0.08, 0.07, 0.15])
})

# Show first 5 rows
orders.head()
```

```python
# Merge temporary product price into orders so sales amount can be created
orders = orders.merge(products[["product_id", "price"]], on="product_id", how="left")

# Create sales amount = quantity * price
orders["sales_amount"] = np.round(orders["quantity"] * orders["price"], 2)

# Create discount percentage
orders["discount_percent"] = np.random.choice([0, 5, 10, 15, 20], n_orders)

# Apply discount to sales amount
orders["final_amount"] = np.round(orders["sales_amount"] * (1 - orders["discount_percent"] / 100), 2)

# Create profit amount as simplified logic
orders["profit_amount"] = np.round(orders["final_amount"] * np.random.uniform(0.08, 0.30, n_orders), 2)

# Drop temporary price if you want orders independent
orders.drop(columns=["price"], inplace=True)

# Add missing values
orders.loc[np.random.choice(orders.index, 100, replace=False), "payment_method"] = np.nan
orders.loc[np.random.choice(orders.index, 90, replace=False), "discount_percent"] = np.nan

# Add inconsistent spacing/casing
orders.loc[np.random.choice(orders.index, 150, replace=False), "order_status"] = " delivered "
orders.loc[np.random.choice(orders.index, 120, replace=False), "payment_method"] = "card"

# Duplicate some rows
order_duplicates = orders.sample(50, random_state=42)

# Create uncleaned orders dataframe
orders_uncleaned = pd.concat([orders, order_duplicates], ignore_index=True)

# Show shape
print("Orders uncleaned shape:", orders_uncleaned.shape)
```

```python
# Positive review examples
positive_reviews = [
    "Excellent product and very useful",
    "Amazing quality and fast delivery",
    "Very happy with this purchase",
    "Good product worth the price",
    "Loved it and will buy again"
]

# Negative review examples
negative_reviews = [
    "Very poor quality and disappointing",
    "Waste of money and bad product",
    "Delivery was late and product was damaged",
    "Not worth the price",
    "Terrible experience with this item"
]

# Total feedback rows
n_feedback = 5000

# Random order ids for feedback
feedback_order_ids = np.random.choice(orders["order_id"], n_feedback)

# Random sentiment labels
sentiment_labels = np.random.choice(["__label__1", "__label__2"], n_feedback, p=[0.35, 0.65])

# Generate review text based on label
review_texts = []
ratings = []

for label in sentiment_labels:
    if label == "__label__2":
        review_texts.append(random.choice(positive_reviews))
        ratings.append(np.random.randint(4, 6))
    else:
        review_texts.append(random.choice(negative_reviews))
        ratings.append(np.random.randint(1, 3))

# Build feedback dataframe
feedback = pd.DataFrame({
    "feedback_id": [f"FDBK_{4000+i}" for i in range(n_feedback)],
    "order_id": feedback_order_ids,
    "raw_sentiment_label": sentiment_labels,
    "review_text": review_texts,
    "rating": ratings,
    "feedback_date": [datetime(2023, 1, 1) + timedelta(days=np.random.randint(0, 730)) for _ in range(n_feedback)]
})

# Show first 5 rows
feedback.head()
```

```python
# Add missing values in review text
feedback.loc[np.random.choice(feedback.index, 100, replace=False), "review_text"] = np.nan

# Add missing values in rating
feedback.loc[np.random.choice(feedback.index, 80, replace=False), "rating"] = np.nan

# Add inconsistent spacing
feedback.loc[np.random.choice(feedback.index, 150, replace=False), "review_text"] = "  amazing quality and fast delivery  "
feedback.loc[np.random.choice(feedback.index, 120, replace=False), "review_text"] = "VERY POOR QUALITY AND DISAPPOINTING"

# Duplicate some rows
feedback_duplicates = feedback.sample(55, random_state=42)

# Create uncleaned feedback dataframe
feedback_uncleaned = pd.concat([feedback, feedback_duplicates], ignore_index=True)

# Show shape
print("Feedback uncleaned shape:", feedback_uncleaned.shape)
```

```python
# Save customer raw table
customers_uncleaned.to_csv("uncleaned_data/customers_uncleaned.csv", index=False)

# Save product raw table
products_uncleaned.to_csv("uncleaned_data/products_uncleaned.csv", index=False)

# Save orders raw table
orders_uncleaned.to_csv("uncleaned_data/orders_uncleaned.csv", index=False)

# Save feedback raw table
feedback_uncleaned.to_csv("uncleaned_data/feedback_uncleaned.csv", index=False)

# Print confirmation
print("All 4 uncleaned tables saved successfully in uncleaned_data folder.")
```
